```python
#File to Load 
df = pd.read_csv("ramen-ratings.csv")
ramen_ratings_df = pd.read_csv("ramen-ratings.csv")
df.head()
```

```python
new_ramen_ratings_df = ramen_ratings_df[['Brand','Variety','Country','Stars']].copy()
new_ramen_ratings_df.head()
```

```python
new_ramen_ratings_df = ramen_ratings_df[["Brand","Variety","Stars"]].copy()
new_ramen_ratings_df.head()
```

```python
df = pd.read_json ("ramen-ratings.json")
ramen_ratings_area_df = pd.read_json("ramen-ratings.json")
ramen_ratings_area_df.head()
```

```python
ramen_ratings_area_df = ramen_ratings_df [["Brand","Country","Stars"]].copy()
ramen_ratings_area_df.head()
```

```python
rds_connection_string = "postgres:Butter12@localhost:5432/ramen_ratings_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
new_ramen_ratings_df.to_sql(name='ramen_ratings', con=engine, if_exists='append', index=False) 
```

```python
ramen_ratings_area_df.to_sql(name='ramen_ratings_area', con=engine, if_exists='append', index=False) 
```
